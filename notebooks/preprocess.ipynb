```python
import fiftyone.zoo as foz

# Download 100 images from the COCO-2017 validation set 
# focusing specifically on the 'person' class
dataset = foz.load_zoo_dataset(
    "coco-2017",
    split="validation",
    label_types=["detections"],
    classes=["person"],
    max_samples=100,
)

# Export these images to a local directory for your noise tests
dataset.export(
    export_dir="./coco_human_subset",
    dataset_type=fo.types.YOLOv8Dataset,
    label_field="ground_truth",
)
```
